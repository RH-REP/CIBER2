```python
import numpy as np
import os
import glob
import spec_reader
import importlib
import matplotlib.pyplot as plt
```

```python
baffle = spec_reader.angle_function.BaffleClass('sp1_base.csv','sp1_base.csv')
cal_file = "specOI/2021-07-07-cossp2-F1.cal"

home_directory = os.environ['HOME']
datadirectory = home_directory + '/spectrometer/spectrometerOI/BB/20201217/1100K/LONG-COSsp1-L1/'

measuredFiles = glob.glob(datadirectory +"long*")
bgFiles = glob.glob(datadirectory +"BG*")
irradiance_1100 = spec_reader.irradiance.IrradianceClass(measuredFiles,bgFiles,baffle = baffle,cal_file=cal_file)
irradiance_1100.show_wave_and_irradiance()

datadirectory = home_directory + '/spectrometer/spectrometerOI/BB/20201217/1200K/LONG-COSsp1-L1/'
measuredFiles = glob.glob(datadirectory +"long*")
bgFiles = glob.glob(datadirectory +"BG*")

irradiance_1200 = spec_reader.irradiance.IrradianceClass(measuredFiles,bgFiles,baffle = baffle,cal_file=cal_file)
irradiance_1200.show_wave_and_irradiance()
```

```python
w,irr_1100 = irradiance_1100.get_wave_and_irradiance()
w,irr_1200 = irradiance_1200.get_wave_and_irradiance()

plt.scatter(w,irr_1100/irr_1200)
plt.ylim(0,1)
plt.xlabel("wavelength [nm]")
plt.ylabel("irradiance ratio 1100/1200")
```
